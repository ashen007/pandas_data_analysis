```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../data/p1_train.csv',index_col='id')
data.head(10)
```

```python
data.isna().sum()
```

```python
plt.figure(figsize=[24,4],dpi=300)
sns.displot(data['loan_amnt'],kde=True)
plt.show()
```

```python
# set_style
plt.figure(figsize=[24,4],dpi=300)
sns.set_style('whitegrid')
sns.lmplot(x='loan_amnt',y='funded_amnt',data=data)
plt.show()
```

```python
plt.figure(figsize=[24,4],dpi=300)
sns.set_style('dark')
sns.lmplot(x='loan_amnt',y='funded_amnt',data=data,fit_reg=False)
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks')
sns.boxplot(data=data[['loan_amnt','funded_amnt']])
plt.show()
```
