```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../data/p1_train.csv',index_col='id')
data.head(10)
```

```python
data.isna().sum()
```

```python
plt.figure(figsize=[24,4],dpi=300)
sns.displot(data['loan_amnt'],kde=True)
plt.show()
```

```python
# set_style
plt.figure(figsize=[24,4],dpi=300)
sns.set_style('whitegrid')
sns.lmplot(x='loan_amnt',y='funded_amnt',data=data)
plt.show()
```

```python
plt.figure(figsize=[24,4],dpi=300)
sns.set_style('dark')
sns.lmplot(x='loan_amnt',y='funded_amnt',data=data,fit_reg=False)
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks')
sns.boxplot(data=data[['loan_amnt','funded_amnt']])
plt.show()
```

```python
# style parameter
sns.axes_style()
```

```python
# can change style using axes style properties
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.scatterplot(x='loan_amnt',y='funded_amnt',data=data)
plt.show()
```

```python
# use preset styles
plt.figure(figsize=[12,12],dpi=300)
sns.set()
sns.set_context('paper')
sns.scatterplot(x='loan_amnt',y='funded_amnt',data=data)
plt.show()
```

```python
# color palette
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.boxplot(data=data[['loan_amnt','funded_amnt']],palette='Blues_d')
plt.show()
```

```python
# custom color palette
plt.figure(figsize=[16,12],dpi=300)
colors = ['#F2E205','#F2CB05','#736002','#403501','#F2B705']
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.boxplot(data=data[['loan_amnt','funded_amnt','funded_amnt_inv','installment']],palette=sns.set_palette(colors))
plt.show()
```

```python
# plot categorical data
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.stripplot(x='grade',y='loan_amnt',data=data,palette='mako')
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.swarmplot(x='grade',y='loan_amnt',data=data,palette='mako')
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.boxplot(x='grade',y='loan_amnt',data=data,palette='mako')
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.violinplot(x='grade',y='loan_amnt',data=data,palette='mako')
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
sns.countplot(x='grade',data=data,palette='mako')
plt.show()
```

```python
plt.figure(figsize=[16,12],dpi=300)
sns.set()
sns.set_style('ticks',{'axes.grid':True})
g = sns.FacetGrid(data,col='grade',hue='sub_grade')
g.map(plt.hist,'int_rate')
g.add_legend()
plt.show()
```
